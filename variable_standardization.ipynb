```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,6,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])

df = pd.DataFrame(
    {"V1" : V1 ,
     "V2" : V2 ,
     "V3" :V3})
df = df.astype(float)
df
```

### Standardizasyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df) #butun degiskenleri standartlastirdi
```

```python
df
```

```python
df.mean()
```

### Normalizasyon

```python
preprocessing.normalize(df) #butun degerleri 0 ile 1 arasinda donusturur
```

### Min-Max Donusumu (istedigimiz 2 deger arasinda donusturme)

```python
scaler = preprocessing.MinMaxScaler(feature_range =(10,20))
```

```python
scaler.fit_transform(df)
```

## Degisken Dönüsümleri

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

### 0-1 Donusumu

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["yeni_sex"] = lbe.fit_transform(df["sex"])
```

```python
df
```

### "1 ve Digerleri(0)" Dönüsümü

```python
df.head()
```

```python
import numpy as np
df["yeni_day"] = np.where(df["day"].str.contains("Sun"),1,0)
#sunlara 1 digerlerine 0 koy
```

```python
df
```
